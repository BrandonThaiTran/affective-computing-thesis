# CHECK AFFECTNET FILES
This code ensure all affectnet images can be opened 

```python
# imports
from __future__ import print_function 
from __future__ import division
import numpy as np

import pandas as pd
import matplotlib.pyplot as plt
import time
import os
import copy
from PIL import Image, ImageFilter

%matplotlib inline
```

```python
# Top level data directory
root_dir = '/home/jupyter/affectnet'

# directory with images
data_dir = root_dir + '/Manually_Annotated_Images'

# directory with csv
csv_dir = root_dir + '/Manually_Annotated_file_lists'
```

```python
# Make dataframes for training and validation 
training_frame = pd.read_csv(csv_dir + '/training.csv')
validation_frame = pd.read_csv(csv_dir + '/validation.csv')
```

```python
n_train = len(training_frame)
n_val = len(validation_frame)
print("Number of training images: {} \nNumber of validation images: {}".format(n_train,n_val))
```

```python
# Check if images are valid

# store bad images 
bad_train_images = [] 
bad_val_images = []

# validation images
print("Checking validation images...")
for idx in range(n_val):
    if idx % 1000 == 0:
        print("Checked {} validation images".format(idx+1))
    img_name = os.path.join(data_dir, validation_frame.iloc[idx, 0])
    try:
        image = Image.open(img_name)
    except:
        print("Image {} is faulty".format(img_name))
        bad_val_images.append(img_name)

# training images
print("Checking training images...")
for idx in range(n_train):
    if idx % 1000 == 0:
        print("Checked {} training images".format(idx+1))
    img_name = os.path.join(data_dir, training_frame.iloc[idx, 0])
    try:
        image = Image.open(img_name)
    except:
        print("Image {} is faulty".format(img_name))
        bad_train_images.append(img_name)
```
